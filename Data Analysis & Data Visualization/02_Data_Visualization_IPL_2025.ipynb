# Data Visualization of IPL AUCTION DATASET

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas
```

```python
file_path = "/content/ipl_2025_auction_players.csv"
df = pandas.read_csv(file_path)
```

```python
# Apply Style to charts
sns.set_style("darkgrid")
```

 # 1. Pie Chart - Players distribution among teams

```python
# This is for missing Data some players do not have a teams
df_filtered = df[df['Team'] != '-']
plt.figure(figsize=(8, 8))
df_filtered['Team'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, cmap='tab10')
plt.title("Players Distribution Among Teams (Excluding Unsold Players)")
plt.ylabel("")
plt.show()
```

# 2. Bar Graph - Number of players per role

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='Type', palette='viridis', order=df['Type'].value_counts().index)
plt.title("Number of Players by Role")
plt.xlabel("Player Role")
plt.ylabel("Count")
plt.show()
```

# 3. Scatter Plot - Player price vs. base price


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Base', y='Sold', hue='Type', style='Type', palette='deep')
plt.title("Base Price vs. Sold Price (Auction Bid Impact)")
plt.xlabel("Base Price (in Cr)")
plt.ylabel("Sold Price (in Cr)")
plt.show()
```
